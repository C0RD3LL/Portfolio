```r
library('openxlsx')
library("dplyr")
library("gridExtra")
library("tibble")
library("tidyverse")
```

```r
x <- read.xlsx("https://raw.githubusercontent.com/PineBiotech/bioinformatics/master/metagenomics1Annotation.xlsx", sheet = 4) 
df <- data.frame(x)

#We can load the data from an excel sheet to our df via xlsx command.

```

```r
# Pie cahrt tells us the proportion of phyla in the sample.
#Make a table that store the percentage of each group!!! 
par(mar = c(4,4,4,4))
pie(table(x[,1]), main = "ERR1250035 by Phylum", cex = 0.6)
```

```r
barplot(table(x[,1]), main = "ERR1250035 by Phylum", xlab = "Bacteria Type",
las = 2, cex.names =0.7, order = TRUE)
```
